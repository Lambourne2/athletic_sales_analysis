```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv('./Resources/athletic_sales_2020.csv')
sales_2021_df = pd.read_csv('./Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
sales_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
sales_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales_df = pd.concat([sales_2020_df, sales_2021_df], ignore_index=True)
```

```python
# Check if any values are null.
sales_df.isnull().sum()
```

```python
# Check the data type of each column
sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
sales_df['invoice_date'] = pd.to_datetime(sales_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
products_sold_df = sales_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
products_sold_df = products_sold_df.rename(columns={'units_sold': 'Total_Products_Sold'})


# Show the top 5 results.
products_sold_df.head()
```

#### Using `pivot_table`

```python
# Using pivot_table
# Show the number products sold for region, state, and city.
products_sold_df = sales_df.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum').reset_index()


# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_df = products_sold_df.rename(columns={'units_sold': 'Total_Products_Sold'})

# Show the top 5 results.
products_sold_df.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Determine which Region had the Most Sales using groupby
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
region_sales_df = sales_df.groupby(['region', 'state', 'city'])['total_sales'].sum().reset_index()
region_sales_df = region_sales_df.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
region_sales_df.head()
```

#### Using `pivot_table`

```python
# Determine which Region had the Most Sales
# Using pivot_table
# Show the total sales for the products sold for each region, state, and city.
region_sales_df = sales_df.pivot_table(index=['region', 'state', 'city'], values='total_sales', aggfunc='sum').reset_index()


# Optional: Rename the "total_sales" column to "Total Sales"
region_sales_df = region_sales_df.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
region_sales_df.head()


# Show the top 5 results.
region_sales_df.sort_values(by='Total Sales', ascending=False).head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Using groupby Determine which Retailer had the Most Sales
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
retailer_sales_df = sales_df.groupby(['retailer', 'region', 'state', 'city'])['total_sales'].sum().reset_index()
retailer_sales_df = retailer_sales_df.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
retailer_sales_df.sort_values(by='Total Sales', ascending=False).head()
```

#### Using `pivot_table`

```python
# Determine which Retailer had the Most Sales using pivot_table
# Show the total sales for the products sold for each retailer, region, state, and city.
retailer_sales_df = sales_df.pivot_table(index=['retailer', 'region', 'state', 'city'], values='total_sales', aggfunc='sum').reset_index()


# Optional: Rename the "total_sales" column to "Total Sales"
retailer_sales_df = retailer_sales_df.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
retailer_sales_df.sort_values(by='Total Sales', ascending=False).head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data. For context product column has Men's Street Footwear and Women's Apparel
womens_footwear_df = sales_df[sales_df['product'] == "Women's Athletic Footwear"]
womens_footwear_df.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_grouped = womens_footwear_df.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum()
womens_footwear_grouped = womens_footwear_grouped.rename('Womens_Footwear_Units_Sold').reset_index()

# Show the top 5 results.
womens_footwear_grouped.sort_values('Womens_Footwear_Units_Sold', ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_pivot = womens_footwear_df.pivot_table(
    index=['retailer', 'region', 'state', 'city'],
    values='units_sold',
    aggfunc='sum'
).rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
womens_footwear_pivot.sort_values('Womens_Footwear_Units_Sold', ascending=False).head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
daily_sales = womens_footwear_df.pivot_table(
    index='invoice_date',
    values='total_sales',
    aggfunc='sum'
).rename(columns={'total_sales': 'Total Sales'})

# Show the table.
daily_sales.sort_values('Total Sales', ascending=False)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = womens_footwear_df.resample('D', on='invoice_date')['total_sales'].sum().reset_index()
daily_sales = daily_sales.rename(columns={'total_sales': 'Total Sales'})
# Sort the resampled pivot table in descending order on "Total Sales".
daily_sales_sorted = daily_sales.sort_values('Total Sales', ascending=False)
daily_sales_sorted.head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_sales = womens_footwear_df.resample('W', on='invoice_date')['total_sales'].sum().reset_index()
weekly_sales = weekly_sales.rename(columns={'total_sales': 'Total Sales'})
weekly_sales_sorted = weekly_sales.sort_values('Total Sales', ascending=False)
weekly_sales_sorted.head(10)
```
